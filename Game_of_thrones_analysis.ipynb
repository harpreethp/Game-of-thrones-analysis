```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
battles = pd.read_csv("battles.csv")
deaths = pd.read_csv("character-deaths.csv")
```

```python
battles
```

```python
deaths
```

```python
battles.describe()
```

```python
battles = battles.drop(['defender_3','defender_4'], axis = 1)
```

```python
deaths.describe()
```

```python
plt.subplot(2,1,1)
sns.countplot(battles['year'])
plt.title('Amount of battles in terms of years')
plt.show()
```

```python
plt.subplot(2,1,2)
sns.countplot(deaths['Death Year'])
plt.title('Amount of death in which year')
plt.show()
```

```python
sns.countplot(deaths['Allegiances'])
plt.xticks(Rotation=80)
plt.title("Which alliance had maximum death")
plt.show()
```

```python
sns.countplot(battles['attacker_outcome'])
plt.title('attackers outcome')
plt.xlabel('Attackers outcome')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,1,1)
sns.countplot(battles['attacker_king'])
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,1,2)
sns.countplot(battles['defender_king'])
```

```python
plt.figure(figsize=(19,10))
plt.subplot(4,1,1)
sns.countplot(battles['attacker_1'],hue=battles['attacker_outcome'])
plt.show()
```

```python
sns.countplot(battles['battle_type'])
plt.title('Type of battle done most')
plt.show()
```

```python
sns.countplot(battles['battle_type'],hue=battles['attacker_outcome'])
plt.title('Result on types of battles')
plt.show()
```

```python
sns.countplot(battles['attacker_king'],hue=battles['battle_type'])
plt.title('Type of battle used by attacker king')
plt.legend(loc = 'upper right')
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.barplot(x = battles['attacker_outcome'],y = battles['attacker_size'])
plt.title('Attackers army size lead to win or lose')
plt.show()
```

```python
sns.barplot(x = battles['attacker_outcome'],y = battles['defender_size'])
plt.title('Defenders army size lead to win or lose')
plt.show()
```

```python
sns.countplot(battles['region'])
plt.title('In which region there are mostly battles done')
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.countplot(battles['region'],hue=battles['attacker_king'])
plt.title('Which king attacked which region')
plt.xticks(rotation = 90)
plt.legend(loc = 'upper right')
plt.show()
```
